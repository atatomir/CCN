```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns; 
```

```python
def constraint(x, y):
    return max(1 - x, y)

axis = np.arange(0., 1., 0.01)
data = [[constraint(x, y) for y in axis] for x in axis]
ax = sns.heatmap(data, vmin=0, vmax=1, xticklabels=False, yticklabels=False)
```
